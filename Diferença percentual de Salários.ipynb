```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df_funcionarios = pd.DataFrame(columns=['Nome', 'Cargo', 'Salario', 'Data_inicio'])

#Aironys ---------------------------------------------------------------------------------------
nome = "Aironys"
cargo = "Analista de Sistemas Sr"
salario = 3500
data_inicio = '2022-09-20'
df_funcionarios.loc[0] = [nome, cargo, salario, data_inicio]

#Lucas -----------------------------------------------------------------------------------------
nome = "Lucas"
cargo = "Analista de Sistemas"
salario = 3000
data_inicio = '2021-01-01'
df_funcionarios.loc[1] = [nome, cargo, salario, data_inicio]


#Fabi ------------------------------------------------------------------------------------------
nome = "Fabi"
cargo = "Analista de Sistemas Sr"
salario = 3800
data_inicio = '2021-01-01'
df_funcionarios.loc[2] = [nome, cargo, salario, data_inicio]

#José -------------------------------------------------------------------------------------------
nome = "José"
cargo = "Estagiário"
salario = 1600
data_inicio = '2021-01-01'
df_funcionarios.loc[3] = [nome, cargo, salario, data_inicio]

#Pedro -------------------------------------------------------------------------------------------
nome = "Pedro"
cargo = "Analista de Sistemas Jr"
salario = 4000
data_inicio = '2021-01-01'
df_funcionarios.loc[4] = [nome, cargo, salario, data_inicio]


#--------------- Criando o DataFrame e Finalizando com gráfico de barras--------------------------

# Criando um DataFrame com os dados dos funcionários
df_funcionarios=pd.DataFrame({
    'Nome':['Aironys', 'Lucas', 'Fabi', 'José', 'Pedro'],
    'Cargo':['Analista de Sistemas Sr', 'Analista de Sistemas', 'Analista de Sistemas Sr', 'Estagiário', 'Analista de Sistemas Jr'],
    'Salario':[3500, 3000, 3800, 1600, 4000]
})


# Calculando a diferença percentual entre os salários
salario_base = df_funcionarios['Salario'][1]
df_funcionarios['Diferenca_percentual'] = (df_funcionarios['Salario'] - salario_base) / salario_base * 100


# Criando o gráfico de barras
fig, ax = plt.subplots()
ax.bar(df_funcionarios['Nome'], df_funcionarios['Diferenca_percentual'], color='green')


# Títulos e os rótulos dos eixos
ax.set_title('Porcentagem de diferença de salário em relação a ' + df_funcionarios['Nome'][4])
ax.set_xlabel('Funcionários')
ax.set_ylabel('Porcentagem de diferença de salário')


#-------------------------Exibição Gráfica de dados-----------------------------------------------------------

# Mostra o gráfico com a variação percentual
plt.show()

#Mostra a Tabela com a variação percentual
df = pd.DataFrame(data=df_funcionarios, index=[0,1,2,3,4])
display(df)
```
